```python

import seaborn as sns
import matplotlib.pyplot as plt

# Apply Seaborn's default theme
sns.set_theme()

# Create a boxplot of Calories Burn for each Exercise
plt.figure(figsize=(10, 5))  # Set the figure size
sns.boxplot(x=data['Exercise'], y=data['Calories Burn'])
plt.title('Calories Burn by Exercise', fontsize=16)
plt.xlabel('Exercise', fontsize=14)
plt.ylabel('Calories Burn', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

```python
import matplotlib.pyplot as plt

exercise_groups = data.groupby('Exercise')['Calories Burn'].apply(list)

exercise_labels = exercise_groups.index
exercise_data = exercise_groups.values

plt.figure(figsize=(10, 5))

plt.boxplot(exercise_data, labels=exercise_labels, notch=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))


plt.title('Calories Burn by Exercise', fontsize=16)
plt.xlabel('Exercise', fontsize=14)
plt.ylabel('Calories Burn', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
