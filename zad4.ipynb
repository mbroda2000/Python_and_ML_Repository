```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data1 = pd.read_csv('/content/drive/MyDrive/Phyton/winequality-red.csv', sep = ';')
data2 = pd.read_csv('/content/drive/MyDrive/Phyton/winequality-white.csv', sep = ';')
```

```python
quality = data1['quality']
data = data1.drop(columns = ['quality'])

from sklearn.preprocessing import (MinMaxScaler,StandardScaler, )

standard_scaler = StandardScaler().set_output(transform="pandas")
data_standardized = standard_scaler.fit_transform(data1)

minmax_scaler = MinMaxScaler().set_output(transform="pandas")
data_normalized = minmax_scaler.fit_transform(data1)

feature1='total sulfur dioxide'
feature2='sulphates'
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(data_standardized[feature1], data_standardized[feature2], label='Standardized data')
plt.scatter(data_normalized[feature1], data_normalized[feature2], label='Min-Max scaled data')

plt.xlabel('Total Sulfur Dioxide')
plt.ylabel('Sulphates')

plt.legend()
plt.grid(True)

plt.show()
```
