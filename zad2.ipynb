```python
import pandas as pd
data = pd.read_csv('drive/MyDrive/exercise_dataset.csv')
```

```python
import matplotlib.pyplot as plt

data['Age'].plot.hist()
plt.title('Age histogram')
plt.xlabel('Age')
plt.ylabel('Quantity')
plt.show()






```

```python
age = data['Age']
plt.hist(age)
plt.title('Age histogram')
plt.xlabel('Age')
plt.ylabel('Quantity')
```

```python
import matplotlib.pyplot as plt


plt.hist(data['Age'])
plt.title('Number of people of a specific age')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

```python
import matplotlib.pyplot as plt


exercise_intensity_counts = data['Exercise Intensity'].value_counts().sort_index()

plt.bar(exercise_intensity_counts.index, exercise_intensity_counts.values)
plt.title('How many peoples do with Intensity Distribution')
plt.xlabel('Exercise Intensity')
plt.ylabel('How many people')
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.scatter(data['Dream Weight'], data['Actual Weight'])
plt.title('Dream Weight vs Actual Weight')
plt.xlabel('Dream Weight')
plt.ylabel('Actual Weight')
plt.show()
```

```python
import matplotlib.pyplot as plt

male_data = data[data['Gender'] == 'Male']['Calories Burn']
female_data = data[data['Gender'] == 'Female']['Calories Burn']

plt.boxplot([male_data, female_data], labels=['Male', 'Female'])
plt.title('Calories Burn by Gender')
plt.xlabel('Gender')
plt.ylabel('Calories Burn')
plt.show()
```
